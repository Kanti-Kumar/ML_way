# How to handle categorical data

Their are 5 techniques to encode or convert categorical features into numbers.

1. Mapping method
2. ordinary encoding
3. label encoding
4. pandas dummies
5. onehot encoding

first 3 will produce the numerical output while the other will produce 1 hot matrix.

```python
import pandas as pd
import seaborn as sns
```

```python
# loading dataset
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

### 1. mapping method

Mapping method is straight forward way to encode categorical features with few categories. Let's apply it to the class feature: It has three categories: Third, First, Second. We create a dictionary whose keys are categories and values are numerics to encode to and then map it to the dataframe

```python
map_dict={
    'First':0,
    'Second':1,
    'Third':2
}
```

```python
titanic['class']=titanic['class'].map(map_dict)
```

```python
titanic.head()
```

```python
titanic['class'].value_counts()
```

### 2. ordinary encoding

```python
# it also convert categorical data into numbers.

from sklearn.preprocessing import OrdinalEncoder

df_data=titanic[['alive','alone']]
encoder=OrdinalEncoder()
df_encoded=encoder.fit_transform(df_data)
```

```python
# the output of the encoder is a numpy array .we convert it back into dataframe.

df_encoded = pd.DataFrame(df_encoded, columns=df_data.columns, index=df_data.index)

#assign the encoded dataframe back to original titanic dataframe
titanic[['alive', 'alone']] = df_encoded

titanic.head()
```

```python
encoder.categories_

# ordinary encoding cannot handle missing values.it gives error.
```

### 3. label encoding

```python
titanic.head()
```

```python
# it also cannot support missing values.

titanic.isnull().sum()
```

```python
# droping null values
titanic_cleaned=titanic.dropna()
```

```python
titanic_cleaned.isnull().sum()
```

```python
from sklearn .preprocessing import LabelEncoder

deck_feat=titanic_cleaned[['deck']]
label_encoder=LabelEncoder()
deck_encoded=label_encoder.fit_transform(deck_feat)
```

```python
# same as that of ordinary encoder ,the output of label encoder is a numpy array.
# convert it to dataframe

deck_encoded_df=pd.DataFrame(deck_encoded,columns=['deck_encoded'],index=deck_feat.index)
titanic_cleaned['deck']=deck_encoded_df
titanic_cleaned.head()
```

```python
label_encoder.classes_
```

```python
titanic_cleaned['deck'].value_counts()
```

### 4. pandas dummies

This is also simple way to handle categorical features.it will create extra features based on the available categories.Let's apply it to the feature 'who'

```python
titanic.head()
```

```python
titanic[['man','women']]=pd.get_dummies(titanic['who'],drop_first=True)
```

```python
titanic[['man', 'women']].head()
```

```python
titanic.head()
```

```python
# converting true/false into 1/0

titanic['man']=titanic['man'].astype(int)
titanic['women']=titanic['women'].astype(int)

titanic.head()
```

### 5. one hot encoding

it will convert a feature into one hot matrix.additional features corresponding to the values of the given categories will be created.

```python
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder()
town_encoded= one_hot.fit_transform(titanic_cleaned[['embark_town']])
```

```python
one_hot.categories_
```

```python
town_encoded
```

```python
# the output of one hot encoder is a sparse matrix. we will need to convert it into numpy array.
town_encoded=town_encoded.toarray()
```

```python
columns=list(one_hot.categories_)
town_df=pd.DataFrame(town_encoded,columns=columns)
town_df.head()
```

```python
len(town_df)
```

```python
len(titanic_cleaned)
```

```python
drop_embark=titanic_cleaned.drop('embark_town',axis=1)
drop_embark[['Cherbourg', 'Queenstown', 'Southampton']] = town_df

drop_embark.head()
```
