#  Data Visualisation using matplotlib

##  1. Basics of matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# to display plots in notebook
%matplotlib inline
```

```python
x_data=[1,2,3,4,5,6,7]
y_data=np.exp(x_data)
# exp() function is used to for do exponential

plt.plot(x_data,y_data)
plt.title('The title')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### Subplots

```python
# plt.subplots(no. of rows,no. of columns,plot number)

plt.subplot(1,3,1)
plt.plot(x_data,y_data)

plt.subplot(1,3,2)
plt.plot(x_data,np.sin(y_data))

plt.subplot(1,3,3)
plt.plot(x_data,np.cos(y_data))
```

### Figure Size

```python
# we can use plt.figure(figsize=(width,height))
plt.figure(figsize=(10,8))
plt.plot(x_data,y_data)
```

```python
# we can also specify dpi(dot per inch) it tells sharpness of image

plt.figure(figsize=(8,5),dpi=100)
plt.plot(x_data,y_data)
```

### Saving Figure

```python
# we can save matplolib figures in diff. formats like in PNG,JPG,PDF,etc

figure=plt.figure(figsize=(8,5))
figure.savefig('Savedfigure.png')

# their is a file Savedfigure.png is saved.
```

### FigureTitle, Axis Labels and Legends

```python
x=np.linspace(0,50,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1)
plt.title('The title')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
# legend

plt.plot(x,y1,label='Y1')
plt.plot(x,y2,label='Y2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc='lower left')
```

### Line colors and styles

```python
plt.figure(figsize=(10,8))

plt.plot(x,y1,color='red')
plt.plot(x,y2,color='y') # yellow
plt.plot(x,y2*2,color='#00FF00') # green color code
plt.plot(x,y2*3,color='0.8') # grayscale
```

```python
# Line style
# '-' = solid line style
#'--' = dashed line
#'-.' = dashed dot line
# ':' =dotted line
# 'b' = blue markers
# 'or' =red circles

# Line marker
# '.'...point marker
# ','...pixel marker
# 'o'...circle marker
# 'v'...triangle_down marker
# '^'...triangle_up marker
# '<'....triangle_left marker
# '>'....triangle_right marker
# '1'....tri_down marker
# '2'....tri_up marker
# '3'....tri_left marker
# '4'....tri_right marker
# 's'....square marker
# 'p'....pentagon marker
# '*'....star marker
# 'h'....hexagon1 marker
# 'H'....hexagon2 marker
# '+'....plus marker
# 'x'....x marker
# 'D'....diamond marker
```

```python
plt.figure(figsize=(10,8))

plt.plot(x,y1,color='red',ls='solid',marker='+',markersize=3)
plt.plot(x,y2,color='b',ls='dotted',marker='D',markersize=4)
```

### Text and Annotation

```python
# add text to the plot
# plt.text(x position,y position,string,color or style)

x=[1,2,3,4,5,6,7,8]
y=np.square(x)
plt.plot(x,y)
plt.text(2,50,'This text is added to plot',color='r')
```

```python
# text annotation

plt.plot(x, y)

plt.annotate('This is the Minimum point', 
             xy=(1,1), 
             xytext=(1.5, 30), arrowprops=dict(facecolor='red'))

plt.annotate('This is the Maximum point', # string to annotate
             xy=(8,64), # the position of x and y
             xytext=(1, 55), # the xy position to add the texts
             bbox=dict(boxstyle='round'), # or boxstyle can be 'round'
             arrowprops=dict(arrowstyle='fancy'))


plt.annotate('Middle, estimated', # string to annotate
             xy=(4.5,22), # the position of x and y
             xytext=(5, 12), # the xy position to add the texts
             bbox=dict(boxstyle='round', fc='none'), # or boxstyle can be 'round'
             arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3))
```

### Object oriented method

```python
# Create empty figure
fig = plt.figure()

# Create axes

ax = plt.axes()

# Plot on that axes

ax.plot(x, y, 'g') # green color

# Add the title 

ax.set_title('This is a title')

# Add X and Y label

ax.set_xlabel('This is X Label')
ax.set_ylabel('This is Y Label')
```

```python
# Create empty figure
#fig = plt.figure()

# Create axes
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot on that axes

axes[0].plot(x, y, 'g') # green color
axes[1].plot(x, y, 'b') # blue color

# Add the title 

axes[0].set_title('This is a title 1')
axes[1].set_title('This is a title 2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('This is Xlabel')
    ax.set_ylabel('This is Y Label')
    ax.set_title('This is a Title')

# Show the figure

fig

# Use plt.tight_layout to adjust the positions of the axes, and labels automatically 
# Comment it and see how bad the figure is
plt.tight_layout()
```

```python
# Create an axis 
# End the function with ; to get rid of unused figure details

ax = plt.axes()

ax.plot(x, y, 'g')
ax.set(title='This is a Title', 
       xlabel='This is xlabel', 
       ylabel='This is ylabel',
       ylim=(-10,10), # This is x limit..Set by ax.set_xlim()
       xlim=(0,25)); # This is y limit..Set by ax.set_ylim() 
```
